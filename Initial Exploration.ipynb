```python
import pandas as pd
import json
```

```python
def load_movies(path):
    """
    This takes TMDB movie data, processes the json columns, and returns a dataframe.
    :param: path of movie csv file
    :return: processed movie file
    """
    df = pd.read_csv(path)
    df['release_date'] = pd.to_datetime(df['release_date']).apply(lambda x: x.date())
    for col in ['genres', 'keywords', 'production_countries', 'production_companies', 'spoken_languages']:
        df[col] = df[col].apply(json.loads)
    return df
```

```python
movie_df = load_movies('./tmdb_5000_movies.csv')
```

```python
def load_credits(path):
    """
    This takes the credits data, processes the json columns, and returns a dataframe.
    :param: Path of the csv file for credits
    :return: A processed credits dataframe.
    """
    df = pd.read_csv(path)
    for col in ['cast', 'crew']:
        df[col] = df[col].apply(json.loads)
    return df
```

```python
credits_df = load_credits("./tmdb_5000_credits.csv")
```
